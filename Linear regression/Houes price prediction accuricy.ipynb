```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Dhaka Rent.csv')
```

```python
df.head()
```

```python
x = df[['area']]
y = df[['rent']]
y.head()
```

```python
plt.scatter(df['area'], df['rent'], marker='o', color='red')
plt.xlabel("Area")
plt.ylabel('Rent')
```

```python
df.area.mean()
```

```python
df.rent.mean()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2500]])
```

```python
df['predict']= reg.predict(x)
df.head()
```

```python
df['residuals'] = df[['rent']]- reg.predict(x)
df.head()
```

```python
plt.scatter(df.area,df.predict,marker='*',color='Black')
plt.xlabel('Rent')
plt.ylabel('Predict')
```

```python
plt.scatter(df['area'], df['rent'], marker='o', color='red')
plt.plot(df.area,df.predict,marker='*',color='Black')
```

```python
reg.score(x,y)
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain, xtest, ytrain, ytest = tts(x,y)
```

```python
reg.fit(xtest,ytest)
```

```python
reg.score(xtest,ytest)
```

```python
reg.score(xtrain,ytrain)
```
