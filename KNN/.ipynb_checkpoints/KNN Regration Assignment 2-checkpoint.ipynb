```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('weight-height.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df.head
```

```python
x = df[['Gender', 'Height']]
y = df['Weight']
```

```python
x.head
```

```python
x.corr()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
```

```python
y_pred = lr_model.predict(x_test)
```

```python
lr_model.predict([[1, 73.847017]])
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2_test = r2_score(y_test, y_pred)
mse
```

```python
print(f"R Squar score for test data: {r2_test}")
```

```python
y_train_pred = lr_model.predict(x_train)

r2_train = r2_score(y_train, y_train_pred)

print(f"R Squar score for training data: {r2_train}")
```

```python
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(x_train, y_train)
```

```python
y_pred_knn = knn_model.predict(x_test)
```

```python
mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)

print(f'KNN Mean Squared Error: {mse_knn}')
print(f'KNN R-squared: {r2_knn}')
```

```python
y_train_pred_knn = knn_model.predict(x_train)
```

```python
r2_knn_train_data = r2_score(y_train, y_train_pred_knn)
print(f"R Squar score for training data: {r2_knn_train_data}")
```

```python
print(f'Linear Regression MSE: {mse}')
print(f'Linear Regression R-squared: {r2_test}\n')

print(f'KNN MSE: {mse_knn}')
print(f'KNN R-squared: {r2_knn}')
```
